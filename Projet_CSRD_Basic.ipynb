### DISTRIBUTION DES INTENSITES ENVIRONNEMENTALES

- COMMENT UNE ENTREPRISE CONSOMME (OU POLLUE) PAR UNITE DE REVENU

- CarbonIntensity = CO₂ / Revenue
- WaterIntensity = Eau / Revenue
- EnergyIntensity = Énergie / Revenue

- Comment les intensités sont réparties dans le dataset:

- quelle valeur revient souvent ?
- est-ce que toutes les entreprises se ressemblent ?
- est-ce que certaines sont très polluantes ?
- est-ce qu'il y a des cas extrêmes ?
- est-ce que les valeurs sont équilibrées ou non ?

- Score Environnment (E) dépend de la qualité des données:

- si données équilibrées = score fiable
- si outliers = score peut être biaisé
- si toutes entreprises = même score = score n'aura pas de sens
- si valeurs trop concentrées = score distinguera mal les entreprises

- En résumé:

- la distribution des intensités environnementales = les valeurs possibles de pollution/ consommation par revenu, et leur fréquence dans le dataset

### CONCLUSION

- “Les intensités environnementales présentent une dispersion suffisante et une absence d’outliers extrêmes, permettant une normalisation MinMax robuste et une construction stable du Score E. Les distributions ne sont ni hyper-concentrées ni fortement asymétriques, ce qui garantit un scoring équilibré entre les entreprises.”

```python
df['CarbonIntensity'] = df['CarbonEmissions'] / df['Revenue']
df['WaterIntensity'] = df['WaterUsage'] / df['Revenue']
df['EnergyIntensity'] = df['EnergyConsumption'] / df['Revenue']
```

### SOCIAL
➡ Forme en cloche, centrée autour de ≈55
➡ Beaucoup d’entreprises ont un score entre 45 et 65
➡ Quelques entreprises vont jusqu’à 80+
➡ Presque aucune est en dessous de 25

👉 Interprétation simple :

La performance sociale varie beaucoup entre les entreprises.

Certaines sont mauvaises (≃30–40), certaines sont excellentes (≃70–85).

Ton Score_S sera très utile pour classer les entreprises car il différencie bien.

---

### GOVERNANCE
➡ Forme aussi en cloche, mais plus resserrée que Social
➡ Centrée autour de ≈45
➡ La plupart des entreprises sont entre 35 et 55
➡ Très peu au-dessus de 60
➡ Quelques rares en dessous de 20

👉 Interprétation simple :

Les niveaux de gouvernance sont moins dispersés.

Les entreprises se ressemblent plus sur cette dimension.

Le Score_G varie moins → il sera plus “doux” dans ton scoring ESG global.

---

### CONCLUSION
Les performances sociales (RH, culture, bien-être…) varient beaucoup

Les structures de gouvernance (CA, règles, transparence) sont souvent plus homogènes

Score_S va fortement influencer ton score ESG global

Score_G va jouer un rôle plus modéré, mais stabilisant

Les deux distributions sont propres → parfait pour normaliser et construire un modèle ESG simple
